## Test .merge on smaller dfs

how : {'left', 'right', 'outer', 'inner'}, default 'inner'

    Type of merge to be performed.

    * left: use only keys from left frame, similar to a SQL left outer join;
      preserve key order.
    * right: use only keys from right frame, similar to a SQL right outer join;
      preserve key order.
    * outer: use union of keys from both frames, similar to a SQL full outer
      join; sort keys lexicographically.
    * inner: use intersection of keys from both frames, similar to a SQL inner
      join; preserve the order of the left keys.

````python
pd.merge(id_topic_head, ft_head, on='ID', how='inner')
#Docstring for merge: The join is done on columns or indexes. If joining columns on columns, the DataFrame indexes *will be ignored*. 
#Otherwise if joining indexes on indexes or indexes on a column or columns, the index will be passed on.

#inner join, only common IDs (order not important, 29555 and 79341 swapped)
#should use this (expect IDs to match and get 2225 rows*)
````
